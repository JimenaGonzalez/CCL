# Calculate NFW profiles with CCL
In this example, we will calculate the 3D and projected 2D NFW profiles for an example cosmology.

```python
import numpy as np
import pyccl as ccl
import pylab as plt
%matplotlib inline
```

### Define cosmology object

```python
cosmo = ccl.Cosmology(Omega_c=0.27, Omega_b=0.045, h=0.67, sigma8=0.83, n_s=0.96)
```

### Define halo properties, scale factor and radii at which to calculate the profile.

```python
concentration = 5
halo_mass = 1E14
overdensity_factor = 200
a = 1
r = np.logspace(-2,2,10)
```

### Calculate and plot 2D projected NFW profile.

```python
sigma_r = ccl.nfw_profile_2d(cosmo, concentration, halo_mass, overdensity_factor, a, r)
```

```python
plt.plot(r, sigma_r)
plt.xlabel(r'$r [Mpc]$')
plt.ylabel(r'$\Sigma(r) [M_{\odot}Mpc^{-2}]$')
plt.loglog();
```

### Calculate and plot 3D NFW profile.

```python
rho_r = ccl.nfw_profile_3d(cosmo, concentration, halo_mass, overdensity_factor, a, r)
```

```python
plt.plot(r, rho_r)
plt.xlabel(r'$r [Mpc]$')
plt.ylabel(r'$\rho(r) [M_{\odot}Mpc^{-3}]$')
plt.loglog();
```

### Calculate and plot 3D Einasto profile.

```python
rho_r = ccl.einasto_profile_3d(cosmo, concentration, halo_mass, overdensity_factor, a, r)
```

```python
plt.plot(r, rho_r)
plt.xlabel(r'$r [Mpc]$')
plt.ylabel(r'$\rho(r) [M_{\odot}Mpc^{-3}]$')
plt.loglog();
```

### Calculate and plot 3D Hernquist profile.

```python
rho_r = ccl.hernquist_profile_3d(cosmo, concentration, halo_mass, overdensity_factor, a, r)
```

```python
plt.plot(r, rho_r)
plt.xlabel(r'$r [Mpc]$')
plt.ylabel(r'$\rho(r) [M_{\odot}Mpc^{-3}]$')
plt.loglog();
```
